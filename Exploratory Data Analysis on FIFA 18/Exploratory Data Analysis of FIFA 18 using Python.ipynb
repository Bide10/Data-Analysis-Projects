## Ask a question about the Data ‚ùì

For any given Overall rating (lets assume 80) does wage vary based on a player's Nationality ?

```python
df[df['Overall']==80].groupby('Nationality')['Wage'].mean().sort_values()
```

```python
ax = df[df['Overall']==80].groupby('Nationality')['Wage'].mean().sort_values().plot(kind = 'barh', figsize =(10,10), title = 'Player Average Wages based on Nationality')
ax.set_xlabel('Average salary in Euros')
plt.show()
```

### Answer: üïµÔ∏è‚Äç‚ôÇÔ∏è
From the above we can answer our question as follows: </br>
Given a fixed overall rating of a player ,say 80 which basically means players of similar merit , we an see a huge <b> nationality biasing</b> as Republic of Ireland is receiving the most salary.  </br>

Not necessarily Rep. of Ireland produces top grade players but given an overall rating of 80 Rep. of Ireland players get and average salary of 110000 euros followed by English players with an average of 99364 euros which is of ~10% difference
